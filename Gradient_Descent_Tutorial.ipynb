```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2+3
print(X,Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Gradient Descent
#initialize x to any random value
x = 0
#learning Rate
lr = 0.1

# since Y = (X-5)^2 + 3
# Gradient dY/dX  = 2*(X-5)
# we will be using formula x = x - lr*(gradient)
# x = x - lr*(2*(X-5))  continuosly done in loop

plt.plot(X,Y)
for i in range(50):
  grad = 2*(x-5)
  x = x - lr*grad;
  y = (x-5)**2 + 3
  plt.scatter(x,y)
  print(x)

plt.show()

```
